```python
from fastai.vision.all import *
from fastai.vision.widgets import *
import gdown
```

```python
# a file
url = "https://drive.google.com/uc?id=1OOfQ6NfC1arpaNQNXtOMzVWeXi0QEiH7"
output = "felinos.zip"
gdown.download(url, output, quiet=False)
!unzip '/content/felinos.zip'
```

```python
path = Path()
learn_inf = load_learner(path/'felinos.pkl', cpu=True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_data_change(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Predicción del modelo: {pred}; Probabilidad: {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change, names=['data'])
```

```python
display(VBox([widgets.Label('Selecciona tu Felino!'), btn_upload, out_pl, lbl_pred]))
```
